```python
import pandas as pd
import numpy as np
```

```python
with open('input.txt', 'r') as f:
    lines = f.readlines()
    
data = []
for line in lines:
    data.append((line[0], line[2]))
    
data
```

```python
shape_points = {
    'A': 1, # opponent rock
    'B': 2, # opponent paper
    'C': 3, # opponent scissors
    'X': 1, # player rock
    'Y': 2, # player paper
    'Z': 3 # player scissors
}
result_points = [0, 3, 6]

def check_win(opponent, player):
    diff = (shape_points[opponent] - shape_points[player]) % 3
    if diff == 0: # draw
        return result_points[1]
    if diff == 1: # lost
        return result_points[0]
    if diff == 2: # won
        return result_points[2]

def compare(opponent, player):
    return shape_points[player] + check_win(opponent, player)

# check given examples
assert compare('A', 'Y') == 8
assert compare('B', 'X') == 1
assert compare('C', 'Z') == 6
```

```python
sum([compare(a,b) for a,b in data])
```

```python
new_tactic = {
    'A': {
        'X': 'Z', # lose
        'Y': 'X', # draw
        'Z': 'Y' # win
    },
    'B': {
        'X': 'X', # lose
        'Y': 'Y', # draw
        'Z': 'Z' # win
    },
    'C': {
        'X': 'Y', # lose
        'Y': 'Z', # draw
        'Z': 'X' # win
    }
}

assert compare('A', new_tactic['A']['Y']) == 4
assert compare('B', new_tactic['B']['X']) == 1
assert compare('C', new_tactic['C']['Z']) == 7
```

```python
sum([compare(a, new_tactic[a][b]) for a,b in data])
```
